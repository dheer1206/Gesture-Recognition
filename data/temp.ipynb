```python
import pandas as pd
import numpy as np
```

```python
l = []
for i in range(1,171) :
    if (i <= 23) :
        l.append(0)
    elif (i <= 45) :
        l.append(1)
    elif (i<=74) :
        l.append(2)
    elif (i <= 103) :
        l.append(3)
    elif (i<=127) :
        l.append(4)
    elif (i<=170) :
        l.append(5)
        
```

```python
l
```

```python
im_id = np.linspace(1,170,170).astype('int32')
```

```python
df = pd.DataFrame(np.array(l),im_id,columns=['Label'])
df.to_csv("/data/out.csv", index_label = ['ImageId'])
```
